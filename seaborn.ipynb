```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('titanic.csv')
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Number of Survivors(Gender)')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Average Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title('Distribution of Fares across passengers')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
sns.violinplot(y='Survived', data=df)
plt.title('Compare the age distribution')
plt.xlabel('Survivors')
plt.ylabel('Non-Survivors(survived)')
plt.show()
```

```python
sns.kdeplot(x='Pclass', y='Survived', data=df)
plt.title('The fare Distribution')
plt.xlabel('survivors')
plt.ylabel('Non-survivors')
plt.show()
```

```python

titanic = sns.load_dataset('titanic')
correlation_matrix = titanic.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
plt.title('Correlation Matrix Heatmap of Titanic Dataset', fontsize=16)
plt.show()
```
